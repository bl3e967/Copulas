```python
# parameters
START_YEAR = 2011
START_MONTH = 1
START_DAY = 1
N_YEARS = 5
N_MONTHS = 0
N_DAYS = 0
RESOLUTION = Resolution.Daily

params = NotebookFuncs.Parameter(RESOLUTION, START_YEAR, START_MONTH, 
                                 START_DAY, N_YEARS, N_MONTHS, N_DAYS)
```
